```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
class PretopologicalSpace:
    def __init__(self, set_elements, relation_matrix):
        self.elements = set_elements
        self.relation_matrix = self.make_idempotent(relation_matrix)

    @staticmethod
    def make_idempotent(matrix):
        """ Update the matrix until it is idempotent (matrix^2 = matrix). """
        while True:
            new_matrix = np.matmul(matrix, matrix)
            new_matrix = (new_matrix > 0).astype(int)  # Convert to 0s and 1s
            if np.array_equal(new_matrix, matrix):
                break
            matrix = new_matrix
        return matrix

    def closure_operator(self, subset):
        """ Compute the closure of a subset using the relation matrix. """
        closure = set(subset)
        for elem in subset:
            related_elements = np.where(self.relation_matrix[elem])[0]
            closure.update(related_elements)
        return closure

    def is_closed(self, subset):
        """ Check if a subset is closed under the closure operator. """
        closure_of_subset = self.closure_operator(subset)
        return set(closure_of_subset) == set(subset)
    
    def find_minimal_closed_subsets(self):
        """ Find all minimal closed subsets. """
        minimal_closed = []
        for element in range(10):  # Assuming elements are integers from 0 to 9
            closed_set = self.closure_operator({element})
            if self.is_closed(closed_set):
                minimal_closed.append(closed_set)
        return minimal_closed

    def find_elementary_closed_subsets(self):
        """ Find all elementary closed subsets. """
        elementary_closed = []
        all_closed_subsets = [self.closure_operator({elem}) for elem in range(10) if self.is_closed({elem})]
        for closed_set in all_closed_subsets:
            if all(not self.is_closed(set(closed_set) - set([elem])) for elem in closed_set):
                elementary_closed.append(closed_set)
        return elementary_closed
```

```python
# Define a 10x10 boolean matrix, reflexive and non-symmetric
relation_matrix = np.eye(10, dtype=int)  # Start with an identity matrix for reflexivity
```

```python
# Example to modify some relations, ensuring non-symmetry
relation_matrix[1, 0] = 1
relation_matrix[2, 0] = 1
relation_matrix[2, 1] = 1
relation_matrix[2, 6] = 1
relation_matrix[3, 1] = 1
relation_matrix[4, 1] = 1
relation_matrix[5, 2] = 1
relation_matrix[6, 5] = 1
relation_matrix[6, 7] = 1
relation_matrix[7, 8] = 1
relation_matrix[8, 7] = 1
relation_matrix[9, 8] = 1
```

```python
relation_matrix=PretopologicalSpace.make_idempotent(relation_matrix)
```

```python
# Example pretopological space
pretop_space = PretopologicalSpace(set(range(10)), relation_matrix)
```

```python
# Find minimal and elementary closed subsets
minimal_closed_subsets = pretop_space.find_minimal_closed_subsets()
elementary_closed_subsets = pretop_space.find_elementary_closed_subsets()
```

```python
print("Minimal Closed Subsets:", minimal_closed_subsets)
```

```python
print("Elementary Closed Subsets:", elementary_closed_subsets)
```

```python
# Graphing the closure relationships
G = nx.DiGraph()
for subset in minimal_closed_subsets + elementary_closed_subsets:
    for elem in subset:
        G.add_node(elem)
        for other in subset:
            if elem != other:
                G.add_edge(elem, other)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title('Pretopological Closure Relations')
plt.show()
```
